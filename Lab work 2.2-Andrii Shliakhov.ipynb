Напишіть функції, що:
* рохраховує мінімальне та максимальне значення
* середнє значення послідовності
* медіану послідовності
* частоту кожної цифри в послідовності
* середнє квадратичне відхилення

Послідовність чисел ганеруємо рандомно

```python
import random
#генеруэмо тисячу чисел в діапазоні від 0 до 10
list_with_random_numbers = [random.randint(0,100) for i in range(1000)]

#print(list_with_random_numbers)
```

### мінімальне значення

```python
def Minimum():
    return min(list_with_random_numbers)
Minimum()
print('мінімальне значення = '+str(Minimum()))
```

### максимальне значення

```python
def Maximum():
    return max(list_with_random_numbers)
Maximum()
print('максимальне значення = '+str(Maximum()))
```

### середнє значення послідовності

```python
def Average():
    return sum(list_with_random_numbers) / len(list_with_random_numbers)
Average()
print('середнє значення послідовності = '+str(round(Average())))
```

### медіана послідовності

```python
def Median():
    list_with_random_numbers.sort()
    half = len(list_with_random_numbers)//2
    m = list_with_random_numbers[half]
    return (m)
Median()
print('медіана послідовності = '+str(Median()))
```

### частота кожної цифри в послідовності

```python
def Freq():
    return {x:list_with_random_numbers.count(x) for x in list_with_random_numbers}
Freq()
```

### середнє квадратичне відхилення

```python
def st_dev():
    from math import sqrt
    return(sqrt(sum((i-Median())**2 for i in list_with_random_numbers)/(len(list_with_random_numbers)-1)))
print('середнє квадратичне відхилення = '+str(round(st_dev())))
```
